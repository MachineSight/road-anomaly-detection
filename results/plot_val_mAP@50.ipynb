```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

files = [
    'results_lr_0.1.csv',
    'results_lr_0.01.csv',
    'results_lr_0.001.csv',
    'results_lr_0.0001.csv',
    'results_lr_0.00001.csv'
]

# learning rates
learning_rates = [0.1, 
                  0.01, 
                  0.001, 
                  0.0001, 
                  0.00001
                 ]

#colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#A133FF']

# Loop through the files and plot mAP@50 for each learning rate
for i, file in enumerate(files):
    # Load the results.csv file
    results = pd.read_csv(file)
    
    plt.plot(results['epoch'],
             results['metrics/mAP50(B)'], 
             label=f'LR={learning_rates[i]}', 
             #color=colors[i]
            )

plt.xlabel('Epoch')
plt.ylabel('mAP@50')
plt.title('Validation mAP@50 vs. Epoch')
plt.legend(title='Learning Rate')
plt.grid(True)

# Save and show the plot
plt.savefig("validation_mAP@0.5.png")
plt.show()
```
